# Average Speed

```python
Speed_mean = df["Speed"].mean()
```

```python
Speed_mean
```

# Function to find low and high speed

```python
def Check_speed(value):
    if(value > Speed_mean):
        return "High_Speed"
    elif(value == Speed_mean):
        return "Neutral_Speed"
    else:
        return "Low_Speed"
```

# Add High_low_speed

```python
df["High_Low_Speed"] = df["Speed"].apply(Check_speed)
```

# Average HP

```python
HP_mean = df["HP"].mean()
```

# Find low and high HP

```python
def Check_HP(value):
    if(value > HP_mean):
        return "High_HP"
    elif(value == HP_mean):
        return "Neutral_HP"
    else:
        return "Low_HP"
```

```python
df["High_Low_HP"] = df["HP"].apply(Check_HP)
```
